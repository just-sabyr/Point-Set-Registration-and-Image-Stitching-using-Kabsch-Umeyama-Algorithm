## Help me create tables in the latex format

### Part A

```python
import numpy as np
```

```python
for sample in ['bottle', 'chair', 'cup']:
    R_p = f"results/{sample}/R.txt"
    T_p = f"results/{sample}/T.txt"
    corr_p = f"results/{sample}/correspondences.png"
    m1_m2_p = f"results/{sample}/mat1_mat2_merged.png"

    R, T = np.loadtxt(R_p), np.loadtxt(T_p)

    print(f"% Sample: {sample}")
    
    # LaTeX for R matrix
    print(f"\\textbf{{Rotation Matrix R for {sample}}}:")
    print("\\[")
    print("\\begin{bmatrix}")
    print(" \\\\\n".join([" & ".join([f"{val:.4f}" for val in row]) for row in R]))
    print("\\end{bmatrix}")
    print("\\]")
    print()

    # LaTeX for T vector
    print(f"\\textbf{{Translation Vector T for {sample}}}:")
    print("\\[")
    print("\\begin{bmatrix}")
    print(" \\\\\n".join([f"{val:.4f}" for val in T]))
    print("\\end{bmatrix}")
    print("\\]")
    print()

    # LaTeX for images
    print(f"\\textbf{{Correspondences Image for {sample}}}:")
    print(f"\\includegraphics[width=0.45\\textwidth]{{{corr_p}}}")
    print()

    print(f"\\textbf{{Merged Mat1 and Mat2 Image for {sample}}}:")
    print(f"\\includegraphics[width=0.45\\textwid th]{{{m1_m2_p}}}")
    print("\n" + "-"*60 + "\n")
```
